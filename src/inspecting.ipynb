```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
filepth = "../public_cases.json"
with open(filepth, 'r') as f:
    data = json.load(f)
df = pd.json_normalize(data)
df.columns = df.columns.str.replace('input.', '', regex=False)
```

```python
df
```

```python
x_axis_column = 'expected_output'
other_columns = df.columns.drop(x_axis_column)

for y_axis_column in other_columns:
    # This single line creates a scatter plot for each pair
    df.plot(
        kind='scatter',
        x=y_axis_column,
        y=x_axis_column,
        title=f'Relationship between {x_axis_column} and {y_axis_column}',
        grid=True, # Adding a grid makes the plot easier to read
        figsize=(8, 5) # You can optionally control the figure size
    )

# 3. Use plt.show() to display all the plots you just created.
# This will open one or more windows showing your graphs.
print("\nDisplaying all plots...")
#plt.show()
```
